```python

def RaDagon(input):
   # temp_arr=
    for i in range(6):
        for Q in range(3):
            input[i,Q]=random.randint(1268,2025)
   # print("Every point have value =\n",i, input[:,:])
    return input  


def newRadaGon(input):
    for Q in range(3):
            input[Q]=random.randint(1268,2025)
   # print("debugaus! Vaihettu arvo = ",input)
    return input

def Calculator(input,Value):
    
    
    Distance=np.sqrt(np.sum((input-Value)**2))
    if math.isnan(Distance):
        print("Nan Value!!",Distance)
        Distance=2000000 
        
        return Distance   
    Distance=int(Distance)
  
   # print(Distance)
    return Distance



def NewMPoint(input,count):
    CalcRes=np.zeros((6,3))
    
    for i in range (6):
    #  print("Input = ")
      for W in range(3):
      #  print("Input = ",input[i,W], "Count = ", count[:,i])
        print("Input:", input)
        print("Count:", count)
        CalcRes[i,W]=int(input[i,W]/count[:,i].item())
       # print(CalcRes)
      
    return(CalcRes)
      
```
